#Задание 1

> С помощью массивов numpy создайте таблицу умножения.



```python
import numpy as np
n = 10
table = np.zeros((n, n), dtype=int)

for i in range(1, n + 1):
    for j in range(1, n + 1):
        table[i-1, j-1] = i * j

print(table)
```

#Задание 2

> Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.



```python
def matrix(N, first_element, difference):
    matrix = np.zeros((N, N), dtype=float)

    for i in range(N):
        matrix[i, i] = first_element + i * difference

    return matrix

N = 3
first_element = 2.0
diff = 3.0
result = matrix(N, first_element, diff)
print(result)
```

# Задание 17

>  Постройте круговую диаграмму для признака Rooms, иллюстрирующую количество квартир в процентах в зависимости от количества комнат. Сделайте сектор с наибольшим числом квартир выдвинутым.



```python
import matplotlib.pyplot as plt

room_counts = df['Rooms'].value_counts()
largest_room_count = room_counts.idxmax()
explode = [0.05 if room == largest_room_count else 0 for room in room_counts.index]

colors = ['#FA8072', '#90EE90', '#F0E68C', '#C2C2F0']

plt.figure(figsize=(6, 6))
plt.pie(room_counts, labels=room_counts.index, autopct='%1.1f%%', startangle=135, explode=explode, colors=colors)
plt.axis('equal')
plt.title('Распределение квартир по количеству комнат')

plt.show()
```

# Задание 18

> Постройте гистограмму по целевой переменной Price. Оцените визуально, по какой цене продаётся наибольшее количество квартир.



```python
plt.figure(figsize=(11, 6))
plt.hist(df['Price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Гистограмма цен на квартиры')
plt.xlabel('Цена квартиры')
plt.ylabel('Количество квартир')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

# Задание 19

> Постройте диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от целевой переменной Price в одной области figure. Оцените визуально, есть ли среди них такие, на которых разброс точек близок к линейной функции.



```python
plt.figure(figsize=(16, 8))

plt.subplot(221)
plt.scatter(df['Rooms'], df['Price'], alpha=0.5)
plt.title('Rooms vs. Price')
plt.xlabel('Rooms')
plt.ylabel('Price')

plt.subplot(222)
plt.scatter(df['Square'], df['Price'], alpha=0.5)
plt.title('Square vs. Price')
plt.xlabel('Square')
plt.ylabel('Price')

plt.subplot(223)
plt.scatter(df['HouseFloor'], df['Price'], alpha=0.5)
plt.title('HouseFloor vs. Price')
plt.xlabel('HouseFloor')
plt.ylabel('Price')

plt.subplot(224)
plt.scatter(df['HouseYear'], df['Price'], alpha=0.5)
plt.title('HouseYear vs. Price')
plt.xlabel('HouseYear')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```
